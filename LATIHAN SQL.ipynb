```python
# import os
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

### 8.How many products in each product line?

```python
# run("SELECT * FROM products")
run("""
SELECT productLine, count(productCode) as JumlahTipeProduk FROM products
    GROUP BY productLine
""")
```

### 9.What is the minimum payment received?

```python
# run("SELECT * FROM payments")
run("""
SELECT * FROM payments
WHERE amount = (SELECT MIN(amount) FROM payments)
""")
# Check
# run("SELECT * FROM payments WHERE amount='615.45'")
```

### 10.List all payments greater than twice the average payment.

```python
# run("SELECT * FROM payments")
run("""
SELECT * FROM payments
WHERE amount >= (2*(SELECT AVG(amount) FROM payments))
""")
# Check
# run("SELECT AVG(amount) FROM payments")
```

### 11. What is the average percentage markup of the MSRP on buyPrice?

```python
# run("SELECT * FROM products")
run("""
SELECT AVG(((MSRP-buyPrice)/MSRP)*100) as RerataProsentaseMurkup FROM products
""")
```

### 12.How many distinct products does ClassicModels sell?

```python
# run("SELECT * FROM orderdetails")
run("""
SELECT count(DISTINCT productCode) as JumlahJenisProdukTerjual FROM orderdetails
""")
```

### 13.Report the name and city of customers who don't have sales representatives?

### 14.What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.

```python
# run("SELECT * FROM employees")
run("""
SELECT employeeNumber, CONCAT(firstName,lastName),extension,email,officeCode,reportsTo,jobTitle FROM employees
WHERE jobTitle <> 'Sales Rep'
""")
```

### 15.Which orders have a value greater than $5,000?

```python
# run("SELECT * FROM orderdetails")
run("""
SELECT *, (quantityOrdered*priceEach) as OrderValue FROM orderdetails
WHERE (quantityOrdered*priceEach) >= 5000
""")
```
