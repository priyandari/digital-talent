```python
import os
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
run("SHOW TABLES")
```

```python
run("SELECT * FROM products")
```

```python
run("SELECT firstName,lastName FROM employees")
```

```python
run("SELECT productName,buyPrice FROM products WHERE buyPrice>50")
```

```python
run("""
SELECT 
    employeeNumber, 
    lastName, 
    firstName
FROM
    employees
WHERE
    lastName LIKE '%on'
""")
```

```python
run("""
SELECT 
    productCode, 
    productName
FROM
    products
WHERE
    productCode LIKE '%\_20%';
""")
```

```python
run("SELECT AVG(buyPrice) AS RataHarga FROM products")
```

```python
run("SELECT productLine, AVG(buyPrice) as RataHarga from products GRUP BY productLine")
```
