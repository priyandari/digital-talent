```python
import os
import pymysql
import pandas as pd

host ='127.0.0.1'
port ='3306'
user ='root'
password=''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
run("SHOW TABLES")
```

```python
run("SELECT * FROM products")
```
