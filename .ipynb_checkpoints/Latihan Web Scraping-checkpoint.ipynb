# Latihan web Scraping

```python
import urllib

r = urllib.request.urlopen('https://tekno.kompas.com/').read()
print(type(r))
print(r[1:1000])
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r,"lxml")
print(type(soup))
print(soup.prettify()[1:1000])
```

```python
links = soup.find_all('a')
print(len(links))

for link in links:
    print('%s: %s'%(link.text.strip(),link["href"]))
    
# Q? Apa maksud pola /read/ dan dari mana kita tahu polanya /read/
```

```python
berita_link = [link for link in links if '/read/' in str(link) and 'javascript:void(0)' not in str(link) and '#' not in str(link)]
# print(len(berita_link))
for link in berita_link:
    print('%s: %s' %(link.text.strip(),link["href"]))
```

## Mendapatkan isi berita

```python
html = urllib.request.urlopen("https://tekno.kompas.com/read/2018/10/23/09230097/demi-reputasi-facebook-dikabarkan-berniat-caplok-perusahaan-sekuriti").read()
soup = BeautifulSoup(html, "lxml")

judul = soup.find("h1",{'class':'read__title'}).text.strip()
isi_berita = soup.find("div","read__content").text.strip()
tanggal = soup.find("div","read__time").text

# Processing tanggal karena aslinya berupa : Kompas.com - 23/10/2018, 09:23 WIB
tanggal = tanggal.strip().split(',')[0].split('-')[1].strip()

print(tanggal)
```

## Gambar

```python
# gambar = soup.find('div',{'class':'photo__wrap'}).find('img')
gambar = soup.find("div","photo").find('img')
print(gambar['src'])

from PIL import Image
# from io import StringIO
from io import BytesIO
import string

r = urllib.request.urlopen(gambar["src"]).read()
print(r)

i = Image.open(BytesIO(r))
exclude = set(string.punctuation)
judul = ''.join(ch for ch in judul if ch not in exclude)

nama_file = "images/" + judul + ".jpg"
i.save(nama_file,'JPEG')
```

# Gabungkan dalam Fungsi

```python
def getBerita(link):
    html = urllib.request.urlopen(link).read()
    soup = BeautifulSoup(html, "lxml")
    
    judul = soup.find("h1",{'class':'read__title'}).text.strip().encode("utf8")
    isi_berita = soup.find("div","read__content").text.strip().encode("utf8")
    tanggal = soup.find("div","read__time").text
    tanggal = tanggal.strip().split(',')[0].split('-')[1].strip()
    gambar = soup.find("div","photo").find('img')
    
    r = urllib.request.urlopen(gambar["src"]).read()
    i = Image.open(BytesIO(r))
    
    exclude = set(string.punctuation)
    judul = ''.join(ch for ch in str(judul) if ch not in exclude)
    
    nama_file = "images/" + judul + ".jpg"
    i.save(nama_file,'JPEG')
    
    return [judul,isi_berita,tanggal]
```

# Dapatkan Semua Data dari Link Berita

```python
berita = []
# berita = berita_link
for link in berita_link:
    print('%s: %s'%(link.text.strip(),link["href"]))
    try:
        berita.append(getBerita(link["href"]))
        
    except:
        pass
    
# print(getBerita("https://tekno.kompas.com/read/2018/10/22/11320037/tahukah-anda-siapa-raksasa-iklan-pemilik-whatsapp-"))
```

```python
#print(berita)
import csv

with open('data_berita.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows(berita)
```
